```python
def get_pk_distribution(degree_list):
    counts = np.bincount(degree_list)
    degrees = np.arange(len(counts))
    nonzero = counts > 0
    pk = counts[nonzero] / sum(counts)
    return degrees[nonzero], pk

def fit_loglog(x, y):
    mask = (x > 0) & (y > 0)
    log_x = np.log10(x[mask])
    log_y = np.log10(y[mask])
    slope, intercept, _, _, _ = linregress(log_x, log_y)
    return slope, intercept
```

```python
def plot_degree_distribution_with_fit(one_def_nx, title=None, save_path=None, ylim=(1e-3, 2e-1)):
    # in-degree, out-degree 리스트 추출
    in_degrees = list(one_def_nx.get('in_degree').values())
    out_degrees = list(one_def_nx.get('out_degree').values())

    # 분포 계산
    in_deg_x, in_deg_pk = get_pk_distribution(in_degrees)
    out_deg_x, out_deg_pk = get_pk_distribution(out_degrees)

    # 회귀 계수 계산
    in_slope, in_intercept = fit_loglog(in_deg_x, in_deg_pk)
    out_slope, out_intercept = fit_loglog(out_deg_x, out_deg_pk)

    # 추세선 생성
    fit_k = np.linspace(1, 100, 100)
    fit_in_pk = 10**in_intercept * fit_k**in_slope
    fit_out_pk = 10**out_intercept * fit_k**out_slope

    # 시각화
    plt.figure(figsize=(6, 6))
    plt.loglog(in_deg_x, in_deg_pk, 'o', label='In-degree', markersize=5)
    plt.loglog(out_deg_x, out_deg_pk, 's', label='Out-degree', markersize=5)
    plt.loglog(fit_k, fit_in_pk, 'k--', label=f'In-degree fit (slope={in_slope:.2f})')
    plt.loglog(fit_k, fit_out_pk, 'k:', label=f'Out-degree fit (slope={out_slope:.2f})')

    plt.xlabel("Degree (k)", fontsize=14)
    plt.ylabel("P(k)", fontsize=14)
    if title:
        plt.title(f"{title}", fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.ylim(ylim)  # ← y축 범위 고정
    plt.tight_layout()

    if save_path is not None:
        plt.savefig(f'{save_path}', dpi=300, bbox_inches='tight')

    plt.show()
```

```python
def plot_strength_histograms(one_def_nx, log_scale=False, title_year=None, save=False, xlim_range=None):
    in_strengths = list(one_def_nx.get('in_strength').values())
    out_strengths = list(one_def_nx.get('out_strength').values())

    plt.figure(figsize=(16, 5))

    # bins 설정
    if xlim_range:
        x_min, x_max = xlim_range
        bins = np.linspace(x_min, x_max, 31)  # 30구간
    else:
        bins = 30

    # in-strength 히스토그램
    plt.subplot(1, 2, 1)
    plt.hist(in_strengths, bins=bins, color='#1f77b4', edgecolor='black')
    if xlim_range:
        plt.xlim(xlim_range)
    plt.title(f"In-Strength Distribution-{title_year} year" if title_year else "In-Strength Distribution", fontsize=14)
    plt.xlabel("In-Strength", fontsize=13)
    plt.ylabel("Node Count", fontsize=13)
    if log_scale:
        plt.yscale('log')

    # out-strength 히스토그램
    plt.subplot(1, 2, 2)
    plt.hist(out_strengths, bins=bins, color='#ff7f0e', edgecolor='black')
    if xlim_range:
        plt.xlim(xlim_range)
    plt.title(f"Out-Strength Distribution-{title_year} year" if title_year else "Out-Strength Distribution", fontsize=14)
    plt.xlabel("Out-Strength", fontsize=13)
    plt.ylabel("Node Count", fontsize=13)
    if log_scale:
        plt.yscale('log')

    plt.tight_layout()

    if save and title_year is not None:
        plt.savefig(f'{pids_info_path}plot_strength_histograms_{title_year}.png', dpi=300, bbox_inches='tight')

    plt.show()
```

```python
def top_n_node_df(one_def_nx, year: int, n: int, ds: str = None, io: str = None, save=False):
    if ds not in ("degree", "strength") and io not in ("in", "out"):
        raise ValueError("You should ds option and io option")

    group = f'{io}_{ds}'
        
    top_n = pd.DataFrame(list(one_def_nx.get(f"{group}").items())[:n]).rename(columns={0:'node_code', 1:'cnt'})
    top_n = pd.merge(top_n, node_pids_info[['node_code', 'Korean', 'English']], on='node_code', how='left')
    
    if save:
        top_n.to_csv(f'{pids_info_path}{group}_{year}.csv', index = False, encoding="utf-8-sig")
    
    print(f"{year}year : {group}")
    print()
    
    return top_n
```

```python
# 그룹 정의
group_definitions = {
    "급성 상기도 감염 질환": {
        "codes": ['J00', 'J02', 'J03', 'J04', 'J06'],
        "color": '#4E79A7',
        "markers": itertools.cycle(['o', 's', '^', 'd', 'v']),
        "styles": itertools.cycle(['-', '--', '-.', ':', '-'])
    },
    "소화기 질환": {
        "codes": ['K29', 'K25'],
        "color": '#59A14F',
        "markers": itertools.cycle(['o', 's']),
        "styles": itertools.cycle(['-', '--'])
    },
    "알레르기성 질환": {
        "codes": ['J30', 'L23'],
        "color": '#F28E2B',
        "markers": itertools.cycle(['o', 's']),
        "styles": itertools.cycle(['-', '--'])
    }
}

# 개별 질병용 색상
individual_color_cycle = itertools.cycle(['#E15759', '#76B7B2', '#B07AA1'])

# 그래프 준비
metrics = ['in_degree', 'out_degree', 'in_strength', 'out_strength']
fig, axes = plt.subplots(2, 2, figsize=(16, 10), sharex=True, sharey=True)
axes = axes.flatten()

all_handles, all_labels = [], []
grouped_codes = set()
style_cache = {}

for i, metric in enumerate(metrics):
    ax = axes[i]
    sub_df = result[result['metric'] == metric]

    for group_name, group_info in group_definitions.items():
        handles = [plt.Line2D([], [], color='white', label=group_name)]
        labels = [group_name]

        for code in group_info['codes']:
            if code not in style_cache:
                style_cache[code] = {
                    'linestyle': next(group_info['styles']),
                    'marker': next(group_info['markers']),
                    'color': group_info['color']
                }
            style = style_cache[code]

            disease_data = sub_df[sub_df['node_code'] == code][['year', 'rank']].sort_values('year')
            full_years = pd.DataFrame({'year': [1, 5, 10]})
            merged = pd.merge(full_years, disease_data, on='year', how='left')

            segments, current = [], []
            for _, row in merged.iterrows():
                if pd.notna(row['rank']):
                    current.append((row['year'], row['rank']))
                else:
                    if current:
                        segments.append(current)
                        current = []
            if current:
                segments.append(current)

            kor = result[result['node_code'] == code]['Korean'].iloc[0]
            label = f"   └── {kor} ({code})"

            for seg in segments:
                x, y = zip(*seg)
                ax.plot(x, y, linestyle=style['linestyle'], marker=style['marker'], color=style['color'],
                        label=label if seg == segments[0] else "")

            # 재등장 텍스트 표기
            prev_rank = None
            for _, row in merged.iterrows():
                if pd.notna(row['rank']) and pd.isna(prev_rank):
                    ax.text(row['year'] - 0.6, row['rank'], code,
                            fontsize=14, verticalalignment='center',
                            alpha=1, fontweight='bold', color=style['color'])
                prev_rank = row['rank']

            handles.append(plt.Line2D([], [], linestyle=style['linestyle'], marker=style['marker'],
                                      color=style['color'], label=label))
            labels.append(label)
            grouped_codes.add(code)

        all_handles.extend(handles)
        all_labels.extend(labels)

    # 그룹 외 개별 질병 처리
    for code in sub_df['node_code'].unique():
        if code in grouped_codes:
            continue

        if code not in style_cache:
            style_cache[code] = {
                'linestyle': '-',
                'marker': 'o',
                'color': next(individual_color_cycle)
            }
        style = style_cache[code]

        disease_data = sub_df[sub_df['node_code'] == code][['year', 'rank']].sort_values('year')
        if disease_data.empty:
            continue

        full_years = pd.DataFrame({'year': [1, 5, 10]})
        merged = pd.merge(full_years, disease_data, on='year', how='left')

        segments, current = [], []
        for _, row in merged.iterrows():
            if pd.notna(row['rank']):
                current.append((row['year'], row['rank']))
            else:
                if current:
                    segments.append(current)
                    current = []
        if current:
            segments.append(current)

        kor = sub_df[sub_df['node_code'] == code]['Korean'].iloc[0]
        label = f"{kor} ({code})"

        for seg in segments:
            x, y = zip(*seg)
            ax.plot(x, y, linestyle=style['linestyle'], marker=style['marker'], color=style['color'],
                    label=label if seg == segments[0] else "")

        prev_rank = None
        for _, row in merged.iterrows():
            if pd.notna(row['rank']) and pd.isna(prev_rank):
                ax.text(row['year'] - 0.6, row['rank'], code,
                        fontsize=14, verticalalignment='center',
                        alpha=1, fontweight='bold', color=style['color'])
            prev_rank = row['rank']

        all_handles.append(plt.Line2D([], [], linestyle=style['linestyle'], marker=style['marker'],
                                      color=style['color'], label=label))
        all_labels.append(label)

    # x/y축 설정
    ax.set_title(metric.replace('_', ' ').title(), fontsize=16)
    ax.set_xticks([1, 5, 10])
    ax.set_xlim(0.3, 10.5)
    ax.invert_yaxis()
    ax.set_ylabel("Rank", fontsize=14)
    ax.set_xlabel("Follow-up period (year)", fontsize=14)
    ax.tick_params(labelsize=13)
    ax.grid(True)
    if i < 2:
        ax.tick_params(labelbottom=True)

# 범례 정리
legend_dict = OrderedDict()
for handle, label in zip(all_handles, all_labels):
    if label not in legend_dict:
        legend_dict[label] = handle

fig.legend(
    legend_dict.values(),
    legend_dict.keys(),
    loc='center left',
    bbox_to_anchor=(1.02, 0.5),
    borderaxespad=0.,
    fontsize=15,
    labelspacing=1.2
)

# fig.suptitle("All Top-6 Diseases Rank Changes Over Time", fontsize=18)
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.savefig(f"{pids_info_path}bump_chart_grouped_final.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
def visualize_for_ppt(year: int, cut_off_per: int = None, cut_off_rr: float = None, vis_opt: bool = False, save_path = None, title_set = None):
    
    cut_result = pd.read_csv(f"{final_results_path}cis_cut_final_result_{year}.csv")
    
    if cut_off_per is not None:
        if 0 <= cut_off_per <= 100:
            cut_result = fun_py.get_top_percent(cut_result, 'rr_values', cut_off_per)
            print(f"rr_value 최소값: {min(cut_result['rr_values'])}")
        else:
            raise ValueError("Error: 'cut_off_per'이 범위를 초과한 값입니다.")
        
    if cut_off_rr is not None:
        cut_result = cut_result[cut_result['rr_values'] >= cut_off_rr]
        
    nx_result =\
    fun_py.compute_network_features(
        filtered_df=cut_result, node_widths=width_dict, node_heights=height_dict
    )
    
    print(Counter(nx_result.get('cluster_assignments').values()))
    
    if vis_opt:
        fun_py.visualize_clusters_from_result(
            result = nx_result,
            layout = 'kamada',
            show_labels = True,
            save_path = save_path,
            title_set = title_set
        )
```

```python
def plot_sex_pie_chart(data, edge=True, save_path=None):
    sex_counts = {
        'Male': data.get('sex_1_counts', 0),
        'Female': data.get('sex_2_counts', 0)
    }
    labels = list(sex_counts.keys())
    sizes = list(sex_counts.values())
    colors = ['#1f77b4', '#ff7f0e']
   
    plt.figure(figsize=(6, 6))
    plt.pie(
        sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        textprops={'fontsize': 14},  # 여기서 폰트 크기를 조절
        colors=colors
    )
    
    if edge:
        title = f"gender distribution: {data['cause_abb']} ➝ {data['outcome_abb']}"
    else:
        title = f"gender distribution: {data['node_code']}"
    
    plt.title(f"{title}", fontsize=14)
    plt.axis('equal')
    
    if save_path:
        plt.savefig(f'{save_path}', dpi=300, bbox_inches='tight')
        
    plt.show()
    
def plot_age_bar_chart(data, edge=True, save_path=None):
    
    age_counts = {int(k.split('_')[1]): v for k, v in data.items() if k.startswith('age_')}
    sorted_ages = dict(sorted(age_counts.items()))

    labels = [str(age) for age in sorted_ages.keys()]
    values = list(sorted_ages.values())
   
    plt.figure(figsize=(10, 5))
    plt.bar(labels, values, color='#1f77b4')
    
    if edge:
        title = f"age distribution: {data['cause_abb']} ➝ {data['outcome_abb']}"
    else:
        title = f"age distribution: {data['node_code']}"
    
    plt.title(f"{title}", fontsize=14)
    plt.xlabel("Age groups")
    plt.ylabel("Number of Patients")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    
    if save_path:
        plt.savefig(f'{save_path}', dpi=300, bbox_inches='tight')
        
    plt.show()
    
def plot_sido_bar_chart(data, edge=True, save_path=None):
    
    sido_counts = {int(k.split('_')[1]): v for k, v in data.items() if k.startswith('sido_')}
    sorted_sido = dict(sorted(sido_counts.items()))

    labels = [str(sido) for sido in sorted_sido.keys()]
    values = list(sorted_sido.values())
   
    plt.figure(figsize=(10, 5))
    plt.bar(labels, values, color='#ff7f0e')
    
    if edge:
        title = f"Region codes distribution: {data['cause_abb']} ➝ {data['outcome_abb']}"
    else:
        title = f"Region codes distribution: {data['node_code']}"
    
    plt.title(f"{title}", fontsize=14)
    plt.xlabel("Region codes")
    plt.ylabel("Number of Patients")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    
    if save_path:
        plt.savefig(f'{save_path}', dpi=300, bbox_inches='tight')
    
    plt.show()
    
def plot_ctrb_bar_chart(data, edge=True, save_path=None):
    
    ctrb_counts = {int(k.split('_')[1]): v for k, v in data.items() if k.startswith('ctrb_')}
    sorted_ctrb = dict(sorted(ctrb_counts.items()))

    labels = [str(ctrb) for ctrb in sorted_ctrb.keys()]
    values = list(sorted_ctrb.values())
   
    plt.figure(figsize=(10, 5))
    plt.bar(labels, values, color='#2ca02c')
    
    if edge:
        title = f"Income levels distribution: {data['cause_abb']} ➝ {data['outcome_abb']}"
    else:
        title = f"Income levels distribution: {data['node_code']}"
    
    plt.title(f"{title}", fontsize=14)
    plt.xlabel("Income levels")
    plt.ylabel("Number of Patients")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    
    if save_path:
        plt.savefig(f'{save_path}', dpi=300, bbox_inches='tight')
    
    plt.show()
    
def plot_sex_age_bar_chart(data, edge=True, save_path=None):
    # 나이별로 성별 count를 분리해서 저장
    sex_age_counts = {'1': {}, '2': {}}

    for key, value in data.items():
        if key.startswith('sex_') and '_age_' in key:
            parts = key.split('_')
            sex = parts[1]
            age = int(parts[3])
            sex_age_counts[sex][age] = value

    # 모든 연령대를 합집합으로 확보하고 정렬
    all_ages = sorted(set(sex_age_counts['1'].keys()) | set(sex_age_counts['2'].keys()))
    
    # 나란히 출력할 수 있도록 데이터 정렬 및 default 0 처리
    male_counts = [sex_age_counts['1'].get(age, 0) for age in all_ages]
    female_counts = [sex_age_counts['2'].get(age, 0) for age in all_ages]

    x = np.arange(len(all_ages))
    width = 0.4

    plt.figure(figsize=(10, 5))
    plt.bar(x - width/2, male_counts, width=width, label='Male', color='#1f77b4')
    plt.bar(x + width/2, female_counts, width=width, label='Female', color='#ff7f0e')

    plt.xlabel("Age group", fontsize=14, labelpad=10)
    plt.ylabel("Number of Patients", fontsize=14, labelpad=10)
    
    if edge:
        title = f"Age Distribution by Sex: {data['cause_abb']} ➝ {data['outcome_abb']}"
    else:
        title = f"Age Distribution by Sex: {data['node_code']}"
    
    plt.title(f"{title}", fontsize=14)
    plt.xticks(x, [str(age) for age in all_ages], rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.legend()
    plt.tight_layout()
    
    if save_path:
        plt.savefig(f'{save_path}', dpi=300, bbox_inches='tight')
    
    plt.show()
    
def plot_sex_sido_bar_chart(data, edge=True, save_path=None):
    
    sex_sido_counts = {'1': {}, '2': {}}

    for key, value in data.items():
        if key.startswith('sex_') and '_sido_' in key:
            parts = key.split('_')
            sex = parts[1]
            sido = int(parts[3])
            sex_sido_counts[sex][sido] = value

   
    all_sidos = sorted(set(sex_sido_counts['1'].keys()) | set(sex_sido_counts['2'].keys()))
    
   
    male_counts = [sex_sido_counts['1'].get(sido, 0) for sido in all_sidos]
    female_counts = [sex_sido_counts['2'].get(sido, 0) for sido in all_sidos]

    x = np.arange(len(all_sidos))
    width = 0.4

    plt.figure(figsize=(10, 5))
    plt.bar(x - width/2, male_counts, width=width, label='Male', color='#1f77b4')
    plt.bar(x + width/2, female_counts, width=width, label='Female', color='#ff7f0e')

    plt.xlabel("Region code", fontsize=14, labelpad=10)
    plt.ylabel("Number of Patients", fontsize=14, labelpad=10)
    
    if edge:
        title = f"Region code Distribution by Sex: {data['cause_abb']} ➝ {data['outcome_abb']}"
    else:
        title = f"Region code Distribution by Sex: {data['node_code']}"
    
    plt.title(f"{title}", fontsize=14)
    plt.xticks(x, [str(sido) for sido in all_sidos], rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.legend()
    plt.tight_layout()
    
    if save_path:
        plt.savefig(f'{save_path}', dpi=300, bbox_inches='tight')
    
    plt.show()
    
def plot_sex_ctrb_bar_chart(data, edge=True, save_path=None):
    
    sex_ctrb_counts = {'1': {}, '2': {}}

    for key, value in data.items():
        if key.startswith('sex_') and '_ctrb_' in key:
            parts = key.split('_')
            sex = parts[1]
            ctrb = int(parts[3])
            sex_ctrb_counts[sex][ctrb] = value

   
    all_ctrbs = sorted(set(sex_ctrb_counts['1'].keys()) | set(sex_ctrb_counts['2'].keys()))
    
   
    male_counts = [sex_ctrb_counts['1'].get(ctrb, 0) for ctrb in all_ctrbs]
    female_counts = [sex_ctrb_counts['2'].get(ctrb, 0) for ctrb in all_ctrbs]

    x = np.arange(len(all_ctrbs))
    width = 0.4

    plt.figure(figsize=(10, 5))
    plt.bar(x - width/2, male_counts, width=width, label='Male', color='#1f77b4')
    plt.bar(x + width/2, female_counts, width=width, label='Female', color='#ff7f0e')

    plt.xlabel("Income levels", fontsize=14, labelpad=10)
    plt.ylabel("Number of Patients", fontsize=14, labelpad=10)
    
    if edge:
        title = f"Income levels Distribution by Sex: {data['cause_abb']} ➝ {data['outcome_abb']}"
    else:
        title = f"Income levels Distribution by Sex: {data['node_code']}"
    
    plt.title(f"{title}", fontsize=14)
    plt.xticks(x, [str(ctrb) for ctrb in all_ctrbs], rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.legend()
    plt.tight_layout()
    
    if save_path:
        plt.savefig(f'{save_path}', dpi=300, bbox_inches='tight')
    
    plt.show()
```

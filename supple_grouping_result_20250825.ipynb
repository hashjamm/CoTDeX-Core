```python
supple_target_code_list = supple_info_2['code'].unique().tolist()
```

```python
approxi_rr_result = edge_stat.loc[(edge_stat['cause_abb'].isin(supple_target_code_list)) | (edge_stat['outcome_abb'].isin(supple_target_code_list))]
```

```python
approxi_rr_result = pd.merge(approxi_rr_result, supple_info_2, how='left', left_on='cause_abb', right_on='code').drop(columns=['code'])
approxi_rr_result.rename(columns={'group_num' : 'cause_group_num'}, inplace=True)

approxi_rr_result = pd.merge(approxi_rr_result, supple_info_2, how='left', left_on='outcome_abb', right_on='code').drop(columns=['code'])
approxi_rr_result.rename(columns={'group_num' : 'outcome_group_num'}, inplace=True)

target_cols = ['cause_group_num', 'outcome_group_num']

# NaN 값을 0으로 채우고, 데이터 타입을 int64로 변환
approxi_rr_result[target_cols] = approxi_rr_result[target_cols].fillna(0).astype('int64')
```
